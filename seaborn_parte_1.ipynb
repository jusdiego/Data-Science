```python
import seaborn as sns
import pandas as pd
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```


O sns.relplot é uma função da biblioteca Seaborn, que é uma biblioteca de visualização de dados baseada no Matplotlib. O sns.relplot é específico para a criação de gráficos de dispersão (scatter plots) com funcionalidades adicionais para facilitar a exploração de relações entre variáveis.

A função sns.relplot é uma interface de alto nível que pode ser usada para criar gráficos de dispersão e/ou de linha com facilidade. Ele permite visualizar a relação entre duas variáveis numéricas e, opcionalmente, incluir uma terceira variável categórica usando diferentes cores ou estilos de linha.

A sintaxe básica do sns.relplot é: <br>
sns.relplot(x='variavel_x', y='variavel_y', data=seu_dataframe, kind='scatter')


```python
sns.relplot(x='SepalLengthCm', y='PetalLengthCm', data=iris);
```

```python
sns.relplot(x='SepalLengthCm', y='PetalLengthCm', data=iris, hue='Species');
```

```python
sns.relplot(x='SepalLengthCm', y='PetalLengthCm', data=iris, hue='Species', style='Species');
```

```python
sns.relplot(x='SepalLengthCm', y='PetalLengthCm', data=iris, hue='Species', size='PetalWidthCm');
```
