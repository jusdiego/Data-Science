```python
# Exemplo do Uso do Apply

# Aqui a função está verificando qual o maior valor do elemento na coluna e marcando ele de amarelo 
def highlight_max(s):
    is_max = s == s.max() # checa esse booleano s == s.max(), e depois guardar em is_max
    return ['background-color: yellow' if v else '' for v in is_max] 
    # O elementeo recebe o 'background-color: yellow' caso ele seja o maior valor

# Aplicar em todo o DataFrame
# Está aplicando apenas no maior elemento da coluna --> .apply
fifa19.head(10).style.apply(highlight_max, subset=['Age', 'Overall', 'Potential'])

```

```python
# Exemplo de alterar valores da coluna 
# A função está tirando do valor o texto 'lbs', e depois transformando x em inteiro
def substitui_lbs(value):
    x = value.replace('lbs', '')
    return int(x)

df['Weight_New'] = df['Weight'].apply(substitui_lbs)
df.head()

# ---------------------------------------------------------
# A função está tirando do valor o texto '€' e 'M' , e depois transformando x em float
def substitui_m(value):
    if value == 'No Release Clause':
        return float('nan')
    value_2 = value.replace('€','')
    if 'M' in value_2:
        value_2 = value_2.replace('M','')
        return float(value_2)*1000000
    elif 'K' in value_2:
        value_2 = value_2.replace('K','')
        return float(value_2)*1000


df['Release Clause_New'] = df['Release Clause'].apply(substitui_m)
df.head()
df['Release Clause_New'].head()

# -------------------------------------------------------
# Alterar o salario para numero

def converte_wage(value):
    value_2 = value.replace('€','')
    if 'M' in value_2:
        value_2 = value_2.replace('M','')
        return float(value_2)*1000000
    elif 'K' in value_2:
        value_2 = value_2.replace('K','')
        return float(value_2)*1000

df['Wage_Numeric'] = df['Wage'].apply(converte_wage)
df.head()

```
