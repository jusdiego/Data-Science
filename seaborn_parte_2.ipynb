```python
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
sns.barplot(x='Species', y='SepalLengthCm', data=iris, hue='Species');
```

```python
sns.barplot(y='Species', x='SepalLengthCm', data=iris, hue='Species');
```

```python
iris['Local_Coleta'] = np.where(iris['Id']%2==0, 'Grande', 'Pequena') 
iris.head()
```

```python
sns.barplot(x='Species', y='SepalLengthCm', data=iris, hue='Local_Coleta');
```

```python
#sns.barplot(x='Species', y='SepalLengthCm', data=iris, ci = 99);
sns.barplot(x='Species', y='SepalLengthCm', data=iris, errorbar=('ci', 99));
# veja mais sobre paletas:
# https://seaborn.pydata.org/tutorial/color_palettes.html
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,10))
sns.barplot(x='Species', y='SepalLengthCm', hue='Local_Coleta', data=iris)
```

```python
# TAREFA DE CASA
ax = sns.barplot(x='Species', y='SepalLengthCm', data=iris, hue='Species');

plt.figure(figsize=(10,8))
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x()+p.get_width()/2.,
            height+.3,
            f'{round(p.get_height(),2)}',
            ha="center",
            rotation=45,
            fontsize=10)
    
ax.set_ylim(top=8) # limitador do eixo y
ax.set_title('Sepal Length por Espcie', fontsize=12); # titulo
ax.set_ylabel("Sepal Length Cm", fontsize=12); # rotulo do eixo y
ax.set_xlabel("Espcie", fontsize=12); # rotulo do eixo x
ax.tick_params(axis='x', labelsize=12, rotation=45) # marcacoes eixo x
ax.spines['top'].set_visible(False) # retirar linha do topo
ax.spines['right'].set_visible(False) # retirar linha lateral direita
```
