```python
# sempres usar o .style para executar uma função

# A função está alterando a cor dos elementos menores que 92 na coluna "Overall"
def low_values_red(value):
    color = 'red' if value < 92 else 'white'
    return f'color: {color}'

# aplicar em todos os elementos --> .map
# df[['coluna_1', 'coluna_2']].head().style.map(funcao)

fifa19[['Overall', 'Potential']].head(10).style.map(low_values_red)
```

```python
def high_values_green(value):
    color = 'green' if value > 92 else 'red'
    return f'color: {color}'

fifa19.head(10).style.map(high_values_green, subset=['Overall', 'Potential'])
```

```python
# Aqui você tem o mesmo resultado, porem vc não está ocultando o resto da DataFrame
# Utilizar o subset=['coluna_1', 'coluna_2', ...] para escolher as colunas onde a função será aplicada
# df.head(10).style.map(funcao, subset=['coluna_1', 'coluna_2'])

fifa19.head(10).style.map(low_values_red, subset=['Overall', 'Potential'])
```

```python
def background_yellow(value):
    return 'background-color: yellow'


fifa19.head(10).style.map(background_yellow, subset=['Age', 'Overall', 'Potential'])
```

```python
# Aqui a função está verificando qual o maior valor do elemento na coluna e marcando ele de amarelo 
def highlight_max(s):
    is_max = s == s.max() # checa esse booleano s == s.max(), e depois guardar em is_max
    return ['background-color: yellow' if v else '' for v in is_max] 
    # O elementeo recebe o 'background-color: yellow' caso ele seja o maior valor

# Aplicar em todo o DataFrame
# Está aplicando apenas no maior elemento da coluna --> .apply
fifa19.head(10).style.apply(highlight_max, subset=['Age', 'Overall', 'Potential'])
```

```python
def no_highlight(value):
    not_is_max = value == value.max()
    return ['background-color: purple' if v else 'background-color: yellow' for v in not_is_max]
 
fifa19.head(10).style.apply(no_highlight, subset=['Age', 'Overall', 'Potential'])
```

```python
# Aqui ele está pegando somente as colunas 'Overall' e 'Potential' e execuntado as função anteriores.

fifa19.head(10).style.map(low_values_red, subset=['Overall', 'Potential']).apply(highlight_max, subset=['Overall', 'Potential'])

fifa19[['Overall', 'Potential']].head(10).style.\
map(low_values_red).\
apply(highlight_max)
```

```python
# Agrupando as medias das colunas 'Age' e 'Overall' com base na nacionalidade
# df.groupby('COLUNA_PRINCIPAL')[['coluna_1', 'coluna_2']].mean().sort_values(by='coluna_n', ascending=False)

df_champs.groupby('Nationality')[['Age', 'Overall']].mean().sort_values(by=['Age', 'Overall'], ascending=False)
```

```python
df_champs.groupby('Nationality')[['Potential', 'Weak Foot']].mean().sort_values(by=['Weak Foot', 'Potential'], ascending=False)
```

```python
import numpy as np
pivot_table_result = pd.pivot_table(df_champs, index='Age', aggfunc=np.mean)
```

```python
# Saber as medias por Nacionalidade das colunas ['Age','Overall']

col = ['Age','Overall']
pd.pivot_table(df_champs, index=['Nationality'], values= col, aggfunc='mean').sort_values(by=col, ascending=False)
```

```python
pd.pivot_table(df_champs, index=['Nationality'], values='Potential', aggfunc='mean').sort_values(by='Potential', ascending=False)
```

```python
# Saber as medias por Nacionalidade das colunas ['Age','Overall','Strength'] por Club 

pd.pivot_table(df_champs, index=['Nationality'], columns=['Club'], values=['Age','Overall','Strength'], aggfunc='mean')
```
