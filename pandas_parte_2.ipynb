```python
# Vendo as informações da coluna

df['Weight'].head()
```

```python
# Se tiver algum missing. Preencher todos os missing por '0lbs'

df['Weight'].fillna('0lbs', inplace=True)
```

```python
# A função está tirando do valor o texto 'lbs', e depois transformando x em inteiro
def substitui_lbs(value):
    x = value.replace('lbs', '')
    return int(x)

df['Weight_New'] = df['Weight'].apply(substitui_lbs)
df.head()
```

```python
df['Release Clause'].head()
```

```python
# A função está tirando do valor o texto '€' e 'M' , e depois transformando x em float
def substitui_m(value):
    if value == 'No Release Clause':
        return float('nan')
    value_2 = value.replace('€','')
    if 'M' in value_2:
        value_2 = value_2.replace('M','')
        return float(value_2)*1000000
    elif 'K' in value_2:
        value_2 = value_2.replace('K','')
        return float(value_2)*1000


df['Release Clause_New'] = df['Release Clause'].apply(substitui_m)
df.head()
df['Release Clause_New'].head()
```

```python
# Alterar o salario para numero

def converte_wage(value):
    value_2 = value.replace('€','')
    if 'M' in value_2:
        value_2 = value_2.replace('M','')
        return float(value_2)*1000000
    elif 'K' in value_2:
        value_2 = value_2.replace('K','')
        return float(value_2)*1000

df['Wage_Numeric'] = df['Wage'].apply(converte_wage)
df.head()
```

```python
# Criando media das características de goleiro 

df['avg_GK'] = (df['GKDiving']+df['GKHandling']+df['GKKicking']+df['GKPositioning']+df['GKReflexes'])/5
df.head()
```

```python
# Cria um campo condicional a outro
df['Age_Class'] = np.where(df['Age']>30, 'Old', 'Young') # import numpy as np
df.head()
```

```python
df['Age_Class2'] = 'Young'
df.loc[df['Age']>30, 'Age_Class2'] = 'Old'
df.head()
```
