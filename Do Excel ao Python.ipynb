### FILTROS

```python
# Filtrando pessoa que não sobreviveram

train[train['Survived']==0]
```

```python
train.query('Age>50')
```

```python
# Filtrando pessoas com mais de 50 anos

train.query('Age>50 & Survived == 1')
```

```python
# Filtrando as mulheres que não sobreviveram

train[(train['Survived']==0) & (train['Sex']=='female')].head()
```

```python
# Criando um df com as colunas ['PassengerId', 'Survived']

cadastro = train[['PassengerId', 'Survived']]
cadastro.head()
```

```python
import numpy as np 

# Criando uma coluna que pega as mulheres que sobreviveram ao acidente e dando o valor de 1, caso ao cotrario 0

train['women_survived'] = np.where((train['Survived']==1) & (train['Sex']=='female'), 0, 1)
train.head()
```
