```python
# EXERCICIO 2

bank['cliente_categoria'] = pd.qcut(bank['balance'], q=[0.8, 0.9, 1], labels=['Gold', 'Premium'])
bank.head()
bank['cliente_categoria'].value_counts()
```

```python
# EXERCICIO 4

# Pegar os 10 maiores clientes em balance
top_10 = bank.nlargest(10, 'balance').sort_values(by='balance', ascending=False)

# Pegar somentes algumas colunas do top_10
df = top_10[['age', 'job', 'marital', 'education', 'default', 'loan', 'duration']]

def attentive(s):
    is_max = s == s.max()
    is_min = s == s.min()
    styles = np.where(is_max, 'background-color: yellow', np.where(is_min, 'background-color: red', ''))
    return styles

df.style.apply(attentive, subset=['duration'])
```

# EXERCICIO 7

```python
brands = pd.read_csv('brands.csv')
brands
```

```python
products = pd.read_csv('products.csv')
products.head()
```

```python
marca_preco = brands.merge(products, on='brand_id')
marca_preco.head()
```

```python
media_por_marca = marca_preco.groupby('brand_name')['list_price'].mean()
media_por_marca
```

# ---------------------------------

# EXERCICIO 8

```python
categorias = pd.qcut(media_por_marca, q=[0, 0.33, 0.66, 1], labels= ["Barato", "Médio", "Caro"])
categorias
```

```python
df_media_categorias = pd.DataFrame({'Média de Preço da Marca': media_por_marca, 'Categoria': categorias})
df_media_categorias
```

```python
bikes = pd.merge(marca_preco, df_media_categorias, left_on='brand_name', right_index=True)
bikes.info()
```
